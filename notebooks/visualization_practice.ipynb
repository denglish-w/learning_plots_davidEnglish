```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

spotify_df = pd.read_csv(r'C:\Users\coldt\Code_You_Projects\module_2\learning_plots_davidEnglish\data\Cleaned_Spotify_2024_Global_Streaming_Data.csv')
spotify_df.head()
```

```python
avg_streams = spotify_df.groupby("Genre")['Total Streams (Millions)'].mean().sort_values(ascending=False)
avg_streams
```

```python
plt.bar(avg_streams.index, avg_streams.values)
plt.show()
```

```python
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel('Genre')
plt.ylabel('Average Total Streams')
plt.title('Average Total Streams By Genre')

plt.show()
```

```python

plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

plt.xlabel('Genre')
plt.ylabel('Average Total Streams')
plt.title('Average Total Streams By Genre')
plt.xticks(rotation=45,ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.bar(avg_streams.index, avg_streams.values)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xlabel('Genre')
plt.ylabel('Average Total Streams')
plt.title('Average Total Streams By Genre')
plt.xticks(rotation=45,ha="right")

plt.tight_layout()
plt.show()
```

```python
avg_artist_streams = spotify_df.groupby('Artist')['Total Streams (Millions)'].mean().sort_values(ascending=False)
avg_artist_streams
```

```python
plt.Figure(figsize=(10,6))
colors = ['#7b66d3']

max_index = avg_artist_streams.idxmax()
highlight_position = avg_artist_streams.index.get_loc(max_index)
colors[highlight_position] = "#5c48b6"

plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xlabel('Genre')
plt.ylabel('Average Total Streams')
plt.title('Average Total Streams By Genre')
plt.xticks(rotation=45,ha="right")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

special_artist = 'Ariana Grande'
colors = ['#7b66d3' if artist == special_artist else 'gray' for artist in avg_artist_streams.index]


plt.bar(avg_artist_streams.index, avg_artist_streams.values, color = colors, width=0.8)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.xlabel('Genre')
plt.ylabel('Average Total Streams')
plt.title('Ariana Grande Ranks Second in total artist streams')
plt.xticks(rotation=45,ha="right")

plt.tight_layout()
plt.show()
```

```python
# Use a scatterplot when you're comparing numbers to numbers

plt.scatter(spotify_df['Streams Last 30 Days (Millions)'], spotify_df['Total Streams (Millions)'])
plt.savefig('../plots/scatter_plot.png')
plt.show()

```

```python
plt.figure(figsize=(10,6))

plt.scatter(
    spotify_df['Streams Last 30 Days (Millions)'], 
    spotify_df['Total Streams (Millions)'],
    color = "#47e261",
    alpha = 0.8,
    s=30
    )
plt.xlabel('Streams this Month')
plt.ylabel('Total Streams')
plt.title('No Apparent Correlation Between Monthly Listens and Total Streams')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)

plt.show()
```

```python
plt.figure(figsize=(10,6))

featured_genre = 'Pop'
colors = ['#47e261' if genre == featured_genre else 'gray' for genre in spotify_df['Genre']]


plt.scatter(
    spotify_df['Streams Last 30 Days (Millions)'], 
    spotify_df['Total Streams (Millions)'],
    color = colors,
    alpha = 0.8,
    s=30
    )
plt.xlabel('Streams this Month')
plt.ylabel('Total Streams')

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.set_title('Do Pop Songs Achieve Higher Monthly Listens or Total Streams?', fontsize = 16, pad=35)

for spine in ax.spines.values():
    spine.set_linewidth(0.25)
    spine.set_alpha(0.5)


plt.tight_layout()
plt.show()
```

```python
yearly_avg = spotify_df.groupby("Release Year")['Total Streams (Millions)'].mean().round(2)
yearly_avg
```

```python
plt.figure(figsize=(10,6))
plt.plot(
    yearly_avg.index,
    yearly_avg.values,
    marker='+',
    linestyle = '--',
    color = "#209434",
    linewidth = (5)
    )

plt.xlabel('Release Year')
plt.ylabel('Average Total Streams')
plt.title('Did Listeners Stream Less in 2022?', fontsize = 18, pad = 20)
plt.ylim(0, 3500)

ax = plt.gca()
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['right'].set_visible(False)

ax.axvline(
    2022,
    color = "#3d7ddd",
)

ax.text(
    2022+0.05,
    2225,
    'Dip in Total Streams',
    color = "#4dc442",
    fontsize=10
)
plt.savefig('../plots/lineGraph.png')
plt.show()
```

```python
numeric_spotify_df = spotify_df.select_dtypes(include=[np.number])
numeric_spotify_df.head()
corr = numeric_spotify_df.corr()

im = plt.imshow(corr, vmin=-1, vmax=1,cmap='coolwarm')
plt.colorbar(im, label='Pearson Correlation')

plt.xticks(range(len(corr.columns)), corr.columns, rotation=45, ha='right', fontsize=8)
plt.yticks(range(len(corr.columns)), corr.columns, )


plt.show()
```
